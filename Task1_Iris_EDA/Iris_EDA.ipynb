# Task 1: Exploring and Visualizing the Iris Dataset

## Problem Statement

The objective of this task is to explore and visualize the Iris dataset
to understand feature distributions, relationships, and potential outliers.

## Goal

- Load and inspect the dataset
- Perform summary statistics
- Visualize trends and distributions
- Extract meaningful insights
  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset Loading
We load the Iris dataset using seaborn.

```python
iris = sns.load_dataset("iris")
iris.head()
```

## Data Inspection
Understanding dataset shape, structure, and summary statistics.

```python
print("Shape:", iris.shape)
print("Columns:", iris.columns)
iris.info()
iris.describe()
```

## Scatter Plot
Visualizing relationship between sepal length and petal length.


```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

Explanation:
Petal length clearly separates species. Sepal length overlaps more.

```python
iris.hist(figsize=(10,8), bins=15)
plt.suptitle("Feature Distributions")
plt.show()
```

Explanation:
Petal features show clearer grouping patterns than sepal features.


```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```

Explanation:
Some mild outliers are visible. Petal width shows strong variation across species.

# Final Insights

Dataset contains 150 samples and 4 numerical features.

No missing values found.

Petal length and petal width are strong distinguishing features.

Sepal measurements overlap more between species.

Dataset is clean and well-structured.
